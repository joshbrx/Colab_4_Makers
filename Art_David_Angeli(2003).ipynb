```python
import numpy as np
import matplotlib.pyplot as plt
import networkx as nx
import random
```

```python
G = nx.DiGraph()
G.add_nodes_from(["$y$","$x$"])
pos = nx.circular_layout(G)

plt.figure(figsize=(15,7))
nx.draw_networkx_nodes(G, pos, node_color = 'tab:green', node_size = 500, alpha = 1,)
labels = nx.draw_networkx_labels(G, pos=pos, font_color="white", font_size=20)
nx.draw_networkx_edges(G, pos, edgelist = [("$x$","$y$")], connectionstyle = f'arc3, rad = 0.3',
                       width = 2, alpha = 1)
nx.draw_networkx_edges(G, pos, edgelist = [("$y$","$x$")], connectionstyle = f'arc3, rad = 0.3',
                       width = 2, alpha = 1)
```
